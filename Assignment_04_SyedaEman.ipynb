# **Question 1**

```python
# Import Packages
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#Loading datset
df = pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
##Create a pairplot to visualize relationships between multiple numerical variables.

# Select numerical columns
numerical_cols = df.select_dtypes(include=[np.number])
# Create a pairplot
sns.pairplot(numerical_cols)
```

```python
## Generate a heatmap to display the correlation matrix of these variables.

# Generate correlation matrix and heatmap
correlation = numerical_cols.corr()
sns.heatmap(correlation, annot=True)
```

```python
## Identify and print pairs of variables with the highest and lowest correlation coefficients

# Identify highest and lowest correlation coefficients
highest_corr = correlation.where(~np.tril(np.ones(correlation.shape)).astype(bool))
lowest_corr = correlation.where(~np.triu(np.ones(correlation.shape)).astype(bool))

# Print highest correlation
print("Highest Correlation (excluding self):")
print(highest_corr.stack().sort_values(ascending=False).head(1))

# Print lowest correlation
print("Lowest Correlation (excluding self):")
print(lowest_corr.stack().sort_values(ascending=True).head(1))
```

```python
#Loading the dataset
df = pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
# Check the first few rows and data types
print(df.head())
print(df.dtypes)
```

```python
# Handle missing values (example)
def handle_missing_values(df):
  # Fill numerical columns with appropriate strategies (e.g., mean, median)
  df['Age'].fillna(df['Age'].mean(), inplace=True)  # Example for Age

  # Handle categorical columns (e.g., most frequent value, imputation)
  df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)  # Example for Embarked

  return df

df = handle_missing_values(df.copy())  # Copy to avoid modifying original DataFrame
```

```python
# Check for and handle outliers (example)
def handle_outliers_iqr(df, column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  # Handle outliers (e.g., capping, winsorization)
  df.loc[df[column] < lower_bound, column] = lower_bound
  df.loc[df[column] > upper_bound, column] = upper_bound

  return df

# Example usage: Handle outliers in 'Fare' column (consider other features)
df = handle_outliers_iqr(df.copy(), 'Fare')

# Alternative methods: winsorization, removal (use cautiously)
```

```python
# Visualize data distributions
df.hist(figsize=(12, 6))
plt.suptitle("Distribution of Titanic Passenger Features")
plt.tight_layout()
plt.show()
```

```python
# Create a new column for age category
def categorize_age(age):
  if age < 18:
    return 'Child'
  elif age >= 65:
    return 'Elderly'
  else:
    return 'Adult'

df['Age_Category'] = df['Age'].apply(categorize_age)
```
